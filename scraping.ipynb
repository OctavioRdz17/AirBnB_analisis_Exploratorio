```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

df = pd.DataFrame(columns=['Neighborhood',"Description", "Price","Total Area","Covered Area","Total Rooms","Bathrooms"])


def scrap_remax(url_list, barrio):
    """Scrap the remax page for a given url and return its content."""

    # Create an empty list to store all of the data.
    data_list = []
    
    for url in url_list:

        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")
        # Encontrar todos los elementos <div> con la clase "mat-ripple info"
        property_elements = soup.find_all("div", class_="mat-ripple info")

        for element in property_elements:
            description = element.find("h2", class_="description ng-star-inserted").text.strip()
            price = element.find("p", id="price").text.strip()
            total_area = element.find("p", {"data-info": "dimensionTotalBuilt"})
            covered_area = element.find("p", {"data-info":"dimensionCovered"})
            total_rooms = element.find("p", {"data-info":"totalRooms"})
            bathrooms = element.find("p", {"data-info":"bathrooms"})
            
            
            data_list.append({
                'Neighborhood': barrio,
                "Description": description,
                "Price": price,
                "Total Area": total_area,
                "Covered Area": covered_area,
                "Total Rooms": total_rooms,
                "Bathrooms": bathrooms
            })

        df_aux = pd.DataFrame(data_list)

    return df_aux


```
